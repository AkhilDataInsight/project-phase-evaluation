```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("your_dataset.csv")  # Replace "your_dataset.csv" with the actual file path

# Exploratory Data Analysis (EDA)
# Perform any necessary data exploration and preprocessing here

# Regression Task: Predict AveragePrice
# Select relevant features for regression
X_reg = data[['feature1', 'feature2', ...]]  # Replace with actual feature names
y_reg = data['AveragePrice']

# Split the data into training and testing sets
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Train a linear regression model
reg_model = LinearRegression()
reg_model.fit(X_reg_train, y_reg_train)

# Make predictions on the test set
y_reg_pred = reg_model.predict(X_reg_test)

# Evaluate the regression model
regression_rmse = mean_squared_error(y_reg_test, y_reg_pred, squared=False)
print(f"Regression RMSE: {regression_rmse}")

# Classification Task: Predict Region
# Select relevant features for classification
X_cls = data[['feature1', 'feature2', ...]]  # Replace with actual feature names
y_cls = data['Region']

# Convert categorical labels to numerical using label encoding or one-hot encoding

# Split the data into training and testing sets
X_cls_train, X_cls_test, y_cls_train, y_cls_test = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42)

# Train a random forest classifier
cls_model = RandomForestClassifier()
cls_model.fit(X_cls_train, y_cls_train)

# Make predictions on the test set
y_cls_pred = cls_model.predict(X_cls_test)

# Evaluate the classification model
classification_accuracy = accuracy_score(y_cls_test, y_cls_pred)
print(f"Classification Accuracy: {classification_accuracy}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load HR data from a CSV file
hr_data = pd.read_csv("hr_data.csv")

# Display the first few rows of the dataset
print(hr_data.head())

# Calculate turnover rate
turnover_rate = (hr_data[hr_data['left'] == 1]['left'].count() / hr_data['left'].count()) * 100
print("Turnover Rate: {:.2f}%".format(turnover_rate))

# Plotting turnover by department
turnover_by_department = hr_data.groupby('department')['left'].mean().sort_values()
turnover_by_department.plot(kind='barh', color='skyblue')
plt.xlabel('Turnover Rate')
plt.ylabel('Department')
plt.title('Turnover Rate by Department')
plt.show()

# Analyze employee satisfaction
satisfaction_mean = hr_data['satisfaction_level'].mean()
print("Average Satisfaction Level: {:.2f}".format(satisfaction_mean))

# Plotting satisfaction level distribution
plt.hist(hr_data['satisfaction_level'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.title('Satisfaction Level Distribution')
plt.show()


```
